# **Ques** In which year most of the games are being published?

```python
df['Year'].value_counts()
```

# **ques** What genre games have been made the most?

# **Ques** Which year had the highest sales worldwide?

```python
new_df=df.groupby(by='Year')['Global_Sales'].sum().sort_values(ascending=False)
print(new_df)
sns.barplot(data=new_df)
plt.xticks(rotation=90)
```

2006 is not in top 5 genre games released but it is in top 5 sales

# **Ques** Which genre game has been released the most in a single year?

```python
new=df.groupby(by='Year')['Genre'].value_counts().unstack().idxmax(axis=1)
new1=df.groupby(by='Year')['Genre'].value_counts().unstack().max(axis=1)
df1=pd.concat([new,new1],axis=1)
df1
```

```python
sns.barplot(data=df1,x=df1.index,y=1,hue=0)
plt.xticks(rotation=90)
```

From 2003 to 2016 maximum released is of action games

# **Ques** Which genre game has sold the most in a single year?

```python
p1=df.groupby(by=['Year','Genre'])['Global_Sales'].sum().unstack().max(axis=1)
p2=df.groupby(by=['Year','Genre'])['Global_Sales'].sum().unstack().idxmax(axis=1)
df2=pd.concat([p2,p1],axis=1)
print(df2)
```

```python
sns.barplot(data=df2,x=df2.index,y=1,hue=0)
plt.xticks(rotation=90)
```

From the last 2 visualizations we get to know that In 2006 most sold is sports game but most released is action games

# **Ques** Which genre game have the highest sale price globally
